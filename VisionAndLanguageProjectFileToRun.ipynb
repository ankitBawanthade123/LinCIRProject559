# **Fetch FashionIQ Images for testing**

```python
import os
import requests
os.chdir('/content/fashioniq-metadata/image_url')



def download_images(file_name, output_dir):
    os.makedirs(output_dir, exist_ok=True)  # Create output directory if it doesn't exist
    successfullImagePath = 0
    with open(file_name, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) == 2:
                image_id, url = parts
                try:
                    response = requests.get(url, stream=True)
                    if response.status_code == 200:
                        image_path = os.path.join(output_dir, f"{image_id}.jpg")
                        with open(image_path, 'wb') as img_file:
                            for chunk in response.iter_content(1024):
                                img_file.write(chunk)
                                successfullImagePath = image_path
                    else:
                        image_path = os.path.join(output_dir, f"{image_id}.jpg")
                        with open(successfullImagePath, 'rb') as src, open(image_path, 'wb') as dst:
                            dst.write(src.read())

                        #print(f"Failed to download {url}, Status Code: {response.status_code}")
                except Exception as e:
                    image_path = os.path.join(output_dir, f"{image_id}.jpg")
                    with open(successfullImagePath, 'rb') as src, open(image_path, 'wb') as dst:
                        dst.write(src.read())


                    #print(f"Error downloading {url}: {e}")

# Step 3: Process all .txt files
txt_files = [file for file in os.listdir() if file.endswith('.txt')]

output_directory = "/content/fashioniq/images"  # Directory to save images

for txt_file in txt_files:
    print(f"Processing {txt_file}...")
    download_images(txt_file, output_directory)

print("Image download complete!")
```
